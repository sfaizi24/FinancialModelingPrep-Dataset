```python
import duckdb
con = duckdb.connect("big.duckdb")  # creates a tiny db file

a_f = '../output/Financial Statements/financials_A-F.csv'
g_l = '../output/Financial Statements/financials_G-L.csv'
m_r = '../output/Financial Statements/financials_M-R.csv'
s_z = '../output/Financial Statements/financials_S-Z.csv'


df = con.sql(f"""
with combined_tables as (
SELECT 
    *
FROM read_csv_auto('{a_f}')
union
select
*
from read_csv_auto('{g_l}')
union
select
*
from read_csv_auto('{m_r}')
union
select
*
from read_csv_auto('{s_z}')
)
select
    symbol
    ,count(*) as a
    ,count(distinct exactDate) as b
    , b - a as c
from combined_tables
group by all
order by c asc

""").df()

df
```

```python
df = con.sql(f"""
with combined_tables as (
SELECT 
    *
    ,'a-f' as source
FROM read_csv_auto('{a_f}')
union
select
*
    ,'g-l' as source
from read_csv_auto('{g_l}')
union
select
*
    ,'m-r' as source
from read_csv_auto('{m_r}')
union
select
*
    ,'s-z' as source
from read_csv_auto('{s_z}')
)
select
distinct symbol
from combined_tables

order by 1
""").df()

df
```
